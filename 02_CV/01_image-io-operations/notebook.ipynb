```python
#reading  videos
cap = cv2.VideoCpture(0) # alternatively 1,-1 are most common ids. One can pass video path too
while(cap.isOpened()): #cap.isOpened() returns True if the cap can be opened
    ret,frame = cap.read()
    #to convert the image into grayscale
    gray_img = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    if ret:
        cv2.imshow('img',frame)
    key_pressed = cv2.waitKey(0)
    if key_pressed == ord('q'):
        break


#releasing the cap 
cap.release()
cv2.destroyAllWindows()
```

```python
#writing videos
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video_writer = cv2.VideoWriter('demo.avi',fourcc, 30.0, (1080,720))
duration  = 50
while(True):
    img = cap.read('lena.png')
    video_writer.write(img)
    if duration > 0:
        video_writer.write(img)
        duration-=1
    else:
        break

video_writer.release()
```
