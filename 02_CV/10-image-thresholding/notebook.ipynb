Image thresholding using OpenCV

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating an image
img = 55*np.ones((1080,1080))
img[540:,540:] = -127
img[500:540,500:540] = 255 
plt.imshow(img,cmap='gray')
```

```python
#thresholding using Binary,Binary inverse and truncated
_, thresholded_img1= cv2.threshold(img,127,255,cv2.THRESH_BINARY)
_, thresholded_img2= cv2.threshold(img,220,255,cv2.THRESH_BINARY_INV)
_, thresholded_img3= cv2.threshold(img,20,255,cv2.THRESH_TOZERO)
_, thresholded_img4= cv2.threshold(img,120,255,cv2.THRESH_TRUNC)
#similar methods of thresholding include THRESH_TO_ZEROINV etc
```

```python
imgs_to_display = [thresholded_img1,thresholded_img2,thresholded_img3,thresholded_img4]
for img in imgs_to_display:
    plt.imshow(img,cmap='gray')
    plt.show()
```

#### Adaptive thresholding
Threshold is calculated for smaller regions and thus is location variant.
> It comes in two flavours :
1. ADAPTIVE_THRESH_MEAN_C
2. ADAPTIVE_THRESH_GAUSSIAN_C

Reference - https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html

```python
img = cv2.imread('sudoku.jpg',0)
plt.imshow(img)
plt.show()
```

```python
simple_thresholded_img = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
mean_thresholded_img = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
gaussian_thresholded_img = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
```

```python
plt.imshow(mean_thresholded_img,cmap='gray')
plt.show()
```

```python
plt.imshow(gaussian_thresholded_img,cmap='gray')
plt.show()
```

#### Remarks
1. Thresholding is useful in finding lesions in images.
2. Adaptive thresholding is very useful where images are illumination sensitive
