#### About
Morphological operations on an image using OpenCV.

> We're going to look at the following operations
1. Image Erosion.
2. Image Dilation.
3. Image Opening which is Erosion followed by Dilation.
4. Image Closing which is Dilation followed by Erosion.
5. Image gradients which is difference between dilation and erosion.
6. Top hat is difference between input image and its opening.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading image
img = cv2.imread('lung_ct.jpeg',0)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
#applying morphological operations
#mask = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,1)
_,mask = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(mask,cmap='gray')
plt.show()
```

```python
#defining kernel
kernel = np.ones((5,5), np.uint8)
#erosion
erosion = cv2.erode(mask,kernel,iterations=1)
plt.imshow(erosion,cmap='gray')
plt.show()
```

```python
#dilation
dilation = cv2.dilate(mask,kernel,iterations=1)
plt.imshow(dilation,cmap='gray')
plt.show()
```

```python
#opening
opening = cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)
plt.imshow(opening,cmap='gray')
plt.show()
```

```python
#closing
closing = cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel)
plt.imshow(closing,cmap='gray')
plt.show()
```

```python
#gradient
grad = cv2.morphologyEx(mask,cv2.MORPH_GRADIENT,kernel)
plt.imshow(grad,cmap='gray')
plt.show()
```

```python
#tophat
tophat = cv2.morphologyEx(mask,cv2.MORPH_TOPHAT,kernel)
plt.imshow(tophat,cmap='gray')
plt.show()
```
