Arithmetic operations on images using OpenCV-Python.

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('demo.jpg',1)
plt.imshow(img)
plt.show()
```

```python
#copying a section and assigning patch to some other place
img_patch = img[90:220,180:370]
plt.imshow(img_patch)
plt.show()
```

```python
#assigning this patch to other area of image
img[170:300,0:190] = img_patch
plt.imshow(img)
plt.show()
```

```python
# adding two images using cv2.add() method
img2 = cv2.imread('demo1.jpg')
#resizing both to common
img2 = cv2.resize(img2,(224,224))
img = cv2.resize(img,(224,224))
added_img = cv2.add(img,img2)
#weighted image method allows us to control proportion
weighted_added_img = cv2.addWeighted(img,0.6, img2,0.4,0)
plt.imshow(added_img)
plt.show()
plt.imshow(weighted_added_img)
plt.show()
```

```python
#similarly substract
sub_img = cv2.subtract(img,img2)
plt.imshow(sub_img)
plt.show()
```

```python
#multiply
mul_img = cv2.multiply(img,img2)
plt.imshow(mul_img)
plt.show()
```

```python
#divide
mul_img = cv2.divide(img,img2)
plt.imshow(mul_img)
plt.show()
```
