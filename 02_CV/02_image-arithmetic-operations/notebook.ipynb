Arithmetic operations in an image.

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('demo.jpg',1)
plt.imshow(img)
plt.show()
```

```python
#copying a section and assigning patch to some other place
img_patch = img[90:220,180:370]
plt.imshow(img_patch)
plt.show()
```

```python
#assigning this patch to other area of image
img[170:300,0:190] = img_patch
plt.imshow(img)
plt.show()
```
