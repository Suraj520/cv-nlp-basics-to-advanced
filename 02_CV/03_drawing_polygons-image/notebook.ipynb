Drawing polygons on Image.

* Images are always loaded in BGR format.

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

sample_img = cv2.imread('sample_img.jpg')
plt.imshow(sample_img)
plt.show()
img = sample_img.copy()
```

```python
# adding a line to it.
sample_img1 = cv2.line(img,(50,0),(50,300),(0,255,0),5 ) #BGR
plt.imshow(sample_img1)
plt.show()
```

```python
#drawing circle
sample_img2 = cv2.circle(img,(300,150),20,(255,255,0),6) #upon passing -1 flag, the image is filled
plt.imshow(sample_img2)
plt.show()
```

```python
#drawing rectangle
sample_img3 = cv2.rectangle(img,(50,0),(550,300),(255,255,0),6) #upon passing -1 flag, the image is filled
plt.imshow(sample_img3)
plt.show()
```

```python
#putting text 
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
sample_img4 = cv2.putText(img,'SampleText',(50,380), font,4, (0,255,0),5, cv2.LINE_8)
plt.imshow(sample_img4)
plt.show()
```
