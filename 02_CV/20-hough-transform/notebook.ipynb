#### About
> Hough Transform
1. It's major application include detection of lines/circles in an image.
2. A line in x,y(cartersian coordinate) space can be expressed as a linear combination of two variables i.e y=mx+c(slope form) or r = xcos*theta) + ysin(theta) in polar coordinates space.
3. The algorithm comprises of the following steps
* Edge detection using an edge detection algorithm.
* Mapping edge points to hough space in an accumulator
* Generate lines of infinite length from accumulator.
* Conversion of infinite lines to finite lines.
4. OpenCV has standard and probabilistic hough lines transform which can be used by cv2.HoughLines and cv2.HoughLinesp. The former being computationally expensive.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


img = cv2.imread('zebra_crossing.jpg')
disp_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#converting to gray
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#detecting edges
unique_pixels = np.unique(gray)
mid_val = unique_pixels[len(unique_pixels)//2]
edges= cv2.Canny(gray,mid_val,255)
plt.imshow(edges,cmap='gray')
```

```python
lines = cv2.HoughLinesP(edges,1,np.pi/180,200,maxLineGap=80) #threshold - higher it is, less lines to be found
```

```python
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv2.line(disp_img,(x1,y1),(x2,y2),(255,0,0),4)

plt.imshow(disp_img)
```
