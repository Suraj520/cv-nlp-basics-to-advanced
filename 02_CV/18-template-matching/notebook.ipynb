#### Template matching using OpenCV

```python
#importing modules
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

template = cv2.imread('template.png',0)
image = cv2.imread('image.jpeg',0)
```

```python
w,h = template.shape[::-1]
#template matching
result = cv2.matchTemplate(image,template,cv2.TM_CCORR_NORMED)
plt.imshow(image)
plt.show()
plt.imshow(result)
plt.show()

```

The brightest region aroiund 150-200 on x axis is our template, Let's filter it

```python
threshold =0.96
loc = np.where(result>=threshold)
for i in zip(*loc[::-1]):
    cv2.rectangle(image,i,(i[0]+w, i[1]+h), (0,0,255),2)


plt.imshow(image)
plt.show()
```
