#### About
1. Fourier transform is used to map spatial domain values to frequency domain values.
2. It assumes that each signal is comprised of sinusoids.
3. Discrete fourier transform is usually applied in digital domains.
4. It is used to design filters after fetching the frequencies. Blurring images are one such application of it.

```python
#importing modules
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Image.png',0)
plt.imshow(img,'gray')
```

```python
freq = np.fft.fft2(img)
fshift = np.fft.fftshift(freq)
magnitude_spectrum = 20* np.log(np.abs(fshift))
print(magnitude_spectrum)
magnitude_spectrum = np.asarray(magnitude_spectrum,dtype=np.uint8)
```

```python
plt.imshow(magnitude_spectrum)
```
